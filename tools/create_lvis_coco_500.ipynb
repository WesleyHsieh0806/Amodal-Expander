```python
import json

gtr_lvis_coco_annotation_path = '/compute/trinity-1-38/chengyeh/lvis/lvis_v1_train+coco_box.json'

with open(gtr_lvis_coco_annotation_path, 'r') as f:
    lvis_coco = json.load(f)

print("Keys in LVIS+COCO: {}".format(lvis_coco.keys()))
```

## Randomly Pick 500 Images

```python
import numpy as np
np.random.seed(0)
print("Original # images: {}".format(len(lvis_coco['images'])))
selected_images = np.random.choice(lvis_coco['images'], size=500, replace=False).tolist()
lvis_coco['images'] = selected_images

selected_image_ids = set([selected_img['id'] for selected_img in selected_images])

assert len(lvis_coco['images']) == len(selected_image_ids)
print("# Selected images: {}".format(len(selected_image_ids)))
print('Examples: {}'.format(lvis_coco['images'][:10]))
```

## Filter Annotations Based on Image Ids

```python
new_anns = []
print("# Original Annotations:{}".format(len(lvis_coco['annotations'])))
for ann in lvis_coco['annotations']:
    if ann['image_id'] in selected_image_ids:
        new_anns.append(ann)

lvis_coco['annotations'] = new_anns
print(lvis_coco['annotations'][:10])
print("# Selected Annotations: {}".format(len(lvis_coco['annotations'])))
```

## Save the New Annotation

```python
print("Save file to {}...".format(gtr_lvis_coco_annotation_path.replace('.json', '_500.json')))
with open(gtr_lvis_coco_annotation_path.replace('.json', '_500.json'), 'w') as f:
    json.dump(lvis_coco, f)
```
